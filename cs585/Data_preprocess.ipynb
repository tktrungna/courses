```python
import json
dataset = '../yelp_dataset/'
business_filename = dataset + 'yelp_academic_dataset_business.json'
user = dataset + 'yelp_academic_dataset_user.json'
review_filename = dataset + 'yelp_academic_dataset_review.json'
tip_filename = dataset + 'yelp_academic_dataset_tip.json'

categories = [u'Restaurants', u'Food', u'Bars', u'Fast Food', u'Pizza', u'Mexican', u'Sandwiches']
```

```python
# reading busines file
i = 0

biz_catg = {}

for l in open(business_filename):
    i += 1
    biz = json.loads(l)
    if i < 2:
        print biz['business_id'], biz['type'], biz['categories']
    if u'Restaurants' not in biz['categories'] and u'Food' not in biz['categories']:
        continue
    for c in biz['categories']:
        biz_catg[c] = biz_catg.get(c,0) + 1
for item in sorted(biz_catg.items(), key=lambda (k, v): v, reverse=True):
    print item
```
